```python
import matplotlib.pyplot as plt
```

```python
processes = ["Steps 0 (initial), 4 (adiabatic compression)", "Step 1, isothermal expansion", 
             "Step 2, adiabatic expansion", "Step 3, isothermal compression", ""]
```

```python
pressures = [49.236,19.6944,3.481510947,8.703777368,49.236]
volumes = [4, 10,28.28427125, 11.3137085, 4]
```

```python
plt.plot(volumes, pressures, "rs-")
plt.xlabel("Volume (L)")
plt.ylabel("Pressure (atm)")
plt.title("P-V Graph")
plt.xlim(0,40)
plt.ylim(0,60)
for i,j in enumerate(pressures):
    plt.text(volumes[i]+0.25, pressures[i]+.5, f"{processes[i]}")
plt.savefig("submission.png")
```
